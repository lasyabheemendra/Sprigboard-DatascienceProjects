```python
import pandas as pd
fires_data = pd.read_csv("fires_clean_data.csv")
```

```python
fires_data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fires_data['STATE'].value_counts()
```

```python
plt.figure(figsize=(15,15))
fires_data['STATE'].value_counts()[:20].plot(kind='barh')
plt.show()
```

```python


plt.figure(figsize=(15,15))
state_count = sns.countplot(x='FIRE_YEAR', data=fires_data,linewidth=5)
state_count.set_xticklabels(state_count.get_xticklabels(), rotation=40, ha="right")
plt.show()
```
